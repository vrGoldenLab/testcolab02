```python
#리스트 예제&튜플

#주사위 빈도 계산하기
import random
a = []
b = []
c = []
d = []
e = []
f = []

for i in range(10000):
    dise = random.randint(1,6)
    if dise == 1 :
        a.append(dise)
    elif dise == 2 :
        b.append(dise) 
    elif dise == 3 :
        c.append(dise) 
    elif dise == 4 :
        d.append(dise) 
    elif dise == 5 :
        e.append(dise)
    elif dise == 6 :
        f.append(dise)  
print('1:',len(a),'2:',len(b),'3:',len(c),'4:',len(d),'5:',len(e),'6:',len(f))
```

```python
#갯수 리스트 만들기
import random
dice_list = [0,0,0,0,0,0]

for i in range(10000):
    dice = random.randint(1,6)
    if dice == 1 :
        dice_list[0] += 1
    elif dice == 2 :
        dice_list[1] += 1 
    elif dice == 3 :
        dice_list[2] += 1 
    elif dice == 4 :
        dice_list[3] += 1 
    elif dice == 5 :
        dice_list[4] += 1
    elif dice == 6 :
        dice_list[5] += 1  
print(dice_list)
```

```python
#count 함수 쓰기 (민준님 코드)
dice = [1,2,3,4,5,6] #주사위를 리스트로 만듦
conse = []
for i in range(10000) :
    conse.append(random.choice(dice))
print(f'''1은 {conse.count(1)}개
2는 {conse.count(2)}개
3은 {conse.count(3)}개
4는 {conse.count(4)}개
5는 {conse.count(5)}개
6은 {conse.count(6)}개
''')
```

```python
#과목 평균 구하기
scores = [[96,84,80], [96,86,76], [76,95,83], [89,96,69], [90,76,91], [82,66,88], [83,86,79], [85,90,83]]

ko = []
en = []
ma =[]
for i in scores:
    ko.append(i[0])
    en.append(i[1])
    ma.append(i[2])

print(ko)
print(en)
print(ma)
print('-'*30)
ko_mean = round(sum(ko)/len(ko),1)
en_mean = round(sum(en)/len(en),1)
ma_mean = round(sum(ma)/len(ma),1)
print(f'국어평균: {ko_mean} , 영어평균: {en_mean}, 수학평균: {ma_mean}')
```

```python
#이중 for문 

kor = []
eng = []
math = []
scores = [[96,84,80], [96,86,76], [76,95,83], [89,96,69], [90,76,91], [82,66,88], [83,86,79], [85,90,83]]
for i in range(3):
    for j in range(len(scores)): #찬민님 코드 수정: 리스트 크기가 커지면 갯수를 알 수 없음
        if i == 0:
            kor.append(scores[j][i])
        elif i == 1:
            eng.append(scores[j][i])
        elif i == 2:
            math.append(scores[j][i])
print(f'국어: {kor}')
print(f'영어: {eng}')
print(f'수어: {math}')
print(f'국어평균: {round(sum(kor)/len(kor),1)}, 영어평균: {round(sum(eng)/len(eng),1)}, 수학평균: {round(sum(math)/len(math),1)}')
```
