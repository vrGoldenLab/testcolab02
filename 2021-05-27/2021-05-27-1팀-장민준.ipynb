## 예제2. 주사위 빈도 계산하기

```python
# random, if문, for문...
import random

dice = list(range(1, 7))
    # 주사위의 눈 개수 설정.
    # range는 끝 값은 들어가지 않는다는 것을 주의.
conse = []

for i in range(10000) :
    conse.append(random.choice(dice))

for i in range(1, len(dice)+1) :
    print(f'{i}은(는) {conse.count(i)} 개')

#print(f'''1은 {conse.count(1)}개
#2는 {conse.count(2)}개
#3은 {conse.count(3)}개
#4는 {conse.count(4)}개
#5는 {conse.count(5)}개
#6은 {conse.count(6)}개
#''')
```

```python
# 선생님의 코드
import random
dice_cnt = [0,0,0,0,0,0]
    # 리스트 초기화.

for i in range(10000) : 
    number = random.randint(1,6)
        # 1~6 랜덤숫자 발생
    if number == 1:
        dice_cnt[0] += 1
    elif number == 2:
        dice_cnt[1] += 1
    elif number == 3:
        dice_cnt[2] += 1
    elif number == 4:
        dice_cnt[3] += 1        
    elif number == 5:
        dice_cnt[4] += 1
    elif number == 6:
        dice_cnt[5] += 1
    # 횟수 카운트. 
        # number 변수를 랜덤으로 1부터 6까지 산출해본다.
        # 그 후에 0부터 5까지 인덱스 넘버를 

print(dice_cnt)
```

## 예제3. 과목별 점수의 평균 구하기

```python
# 1팀 장민준 코드
# if문 사용해서 각각의 length가 0이 아닐때 평균값을 구하면 됨.

score_full = [[96,84,80], [96,86,76], [76,95,83], [89,96,69], [90,76,91], [82,66,88], [83,86,79], [85,90,83]]

# 국어
# for i in range(1, len(score_full)+1) : 
lang = []
for i in score_full : 
    lang.append(i[0])
lang_average = round(sum(lang) / len(lang), 1)

# 영어
eng = []
for i in score_full : 
    eng.append(i[1])
eng_average = round(sum(eng) / len(eng), 1)

# 수학
math = []
for i in score_full : 
    math.append(i[2])
math_average = round(sum(math) / len(math), 1)

print(f'''국어: {lang}
영어: {eng}
수학: {math}
국어평균:{lang_average}, 영어평균:{eng_average}, 수학평균:{math_average}''')
```

```python
# 선생님의 코드

scores = [[96,84,80], [96,86,76], [76,95,83], [89,96,69], [90,76,91], [82,66,88], [83,86,79], [85,90,83]]
koli = []
enli = []
mtli = []

for score in scores :           # list를 범위로 주면 하나하나 항목값을 index 값으로 읽어온다.
    koli.append(score[0])
    enli.append(score[1])
    mtli.append(score[2])

print(koli)
print(enli)
print(mtli)
ave1 = round(sum(koli) / len(koli), 1)
ave2 = round(sum(enli) / len(enli), 1)
ave3 = round(sum(mtli) / len(mtli), 1)

print(f'국어평균: {ave1}, 영어평균: {ave2}, 수학평균: {ave3}')
```

```python
# 3팀 전승혜님 코드

k = []
e = []
m = []
scores = [[96,84,80],[96,86,76],[76,95,83],[76,95,69],[90,76,91],[82,66,88],[83,86,79],[85,90,83]]

for x in range(len(scores)):
    k.append(scores[x][0])
for x in range(len(scores)):
    e.append(scores[x][1])
for x in range(len(scores)):
    m.append(scores[x][2])

print(f'국어: {k}')
print(f'영어: {e}')
print(f'수학: {m}')

if len(k) != 0:
    average1 = round(sum(k)/len(k),1)
else:
    print('평균을 구할 수 없습니다.')
if len(e) != 0:
    average2 = round(sum(e)/len(e),1)
else:
    print('평균을 구할 수 없습니다.')
if len(m) != 0:
    average3 = round(sum(m)/len(m),1)
else:
    print('평균을 구할 수 없습니다.')

print(f'국어 평균: {average1}, 영어 평균: {average2}, 수학 평균: {average3}')
```

## 예제4. 마일리지 관리
- max value로 max key 불러오기

```python
# 내 코드.

mil = {
    'Kim':12000,
    'Lee':11000,
    'Han':3000,
    'Hong':5000,
    'Hwang':18000
}
print('2021.05.26 고객 마일리지 리스트: ', mil)

mil['Han'] = 5000
mil['Jang'] = 7000

print('2021.05.27 고객 마일리지 리스트: ', mil)

#====================================================================
# 마일리지 최고점수 고객 산출
#현재 답안에서는 마지막에 결과값으로 가장 높은 마일리지 점수의 값을 가져오고 있습니다!
#그런데 저는 마일리지 점수의 값이 아니라 고객의 이름을 띄워주는걸로 만들고 싶어요.
#가장 마일리지 점수가 높은 사람을 찾아낸 뒤, 그 사람의 이름을 출력해주는 코드를 만들고자 합니다.

#     1. 가장 높은 마일리지 점수(밸류값) 확인. 아마도 max 통해서.
#     2. max 밸류값을 받아서 그 밸류값을 갖는 키값 찾기.
#     3. 그 키값 출력

#이렇게 해보고 싶은데, 방법이 있을까요?
# => 답변완료

print(list(mil.keys()))
print( list(mil.values()).index( max(mil.values()) ) )

print( list( mil.keys() ) [ list(mil.values()).index( max(mil.values()) ) ] )
    # 장예훈 코드메이트님의 코드.
    # 일단 딕셔너리의 키값들을 list화 시킨다. list(mil.keys())
    # list[인덱스 번호]로 해당 위치에 있는 요소를 불러올 수 있다는 점을 다시 떠올리자.
    # [인덱스 번호] 에...
        # 밸류값(마일리지 점수)으로 리스트를 또 하나 더 만든다.
        # 거기서 인덱싱을 한다. 딕셔너리의 요소의 위치는 제멋대로라고 하더라도, 한번 설정된걸 같은 줄에서 똑같이 참조하고 있으므로 요소가 달라질 염려가 없다.
            # 딕셔너리의 밸류값을 기준으로 max는 어떤것인지 산출. 그게 밸류값 리스트에서 어느 인덱스 넘버로 있는지 인하는것.
    # 이제 완성. max value 값으로 max key 값을 알아내는 코드가 되었다.
```

## 예제5. 단어장 만들기 (어려움)


```python
# 내 코드

dict_voca = {}

#=====================================================
# 1단계. 유저에게 딕셔너리 입력하게 하기
while True:
    eng = input('eng: ')
    if eng=='':
        break       
        # 아무것도 입력하지 않고 공백으로 엔터만 치면, 반복문이 끝나게 해주는 조건문.
    kor = input('kor: ')
    dict_voca[eng] = kor
    # 영단어가 key, 한글 뜻이 value.
print('단어장: ', dict_voca)

#=====================================================
# 2단계. 단어테스트 구현
#리스트에 입력된 순서대로 테스트 실시
    #맞으면 1점씩.
#맞은 개수 계산
    #for문으로 딕셔너리 전체를 순회. 영단어를 제시하면 뜻을 쓰도록 하기.
    #len(딕셔너리명)
#테스트가 끝나면 점수 출력

eng_list = list( dict_voca.keys() )

score = 0

for i in eng_list :
    korean = input(f'{i}의 뜻은?: ')
    if korean == dict_voca[i] :
        score += 1

print(f'맞은갯수/전체단어수: {score}/{len(dict_voca)}')
```
