```python
#주사위 빈도 계산하기
import random
dice_cnt = [0,0,0,0,0,0]           #리스트 초기화

for i in range(10000):             #10000번 반복
    number = random.randint(1,6)   #1~6번 랜덤숫자 발생
    if number == 1:
        dice_cnt[0]+=1
    elif number == 2:
        dice_cnt[1]+=1
    elif number == 3:
        dice_cnt[2]+=1             #횟수 카운트
    elif number == 4:
        dice_cnt[3]+=1
    elif number == 5:
        dice_cnt[4]+=1
    elif number == 6:
        dice_cnt[5]+=1

print(dice_cnt)                    #결과 출력
```

```python
#과목별 점수의 평균 구하기
 scores = [[96,84,80],[96,86,76],[76,95,69],[90,76,91],[82,66,88],[83,86,79],[85,90,83]]
 
 kor = []
 eng = []
 math = []

 for score in scores:
     kor.append(score[0])
     eng.append(score[1])
     math.append(score[2])
print(kor)
print(eng)
print(math)
ave1 = round(sum(kor)/len(kor),1)
ave2 = round(sum(eng)/len(eng),1)
ave3 = round(sum(math)/len(math),1)

print(f'국어평균:{ave1} ,영어평균: {ave2} ,수학평균: {ave3} ')
```

```python
#딕셔너리>데이터 모두 출력하기
  ##dict_keys, dict_values, dict_items는 리스트로 변환하지 않더라도 기본적인 반복구문을 실행할 수 있다.

dict_scores = {
    '홍길동':80,
    '공유':90,
    '현빈':100
}

for key in dict_scores.keys():
    print(key, dict_scores[key])
```

```python
#딕셔너리>데이터 정렬하여 모두 출력하기
 ##키(key)를 정렬하여 정렬된 키(key) 순서대로 출력한다.

dict_scores = {
    '홍길동':80,
    '공유':90,
    '현빈':100
}

# 딕셔너리의 키를 정렬하여 리스트로 받음
sorted_keys = sorted(dict_scores.keys())
print(sorted_keys)

#정렬된 키 리스트 순서로 반복
for key in sorted_keys:
    print(key, dict_scores[key])
```

```python
dict_scores = {
    '홍길동':80,
    '공유':90,
    '현빈':100
}

# 딕셔너리의 키를 정렬하여
# 튜플(키,값)으로 받음
sorted_items = sorted(dict_scores.items())
print(sorted_items)

# 정렬된 키 리스트 순서로 반복
for key,value in sorted_items:
    print(key,value)
```

```python
#예제 단어장 만들기
  ##엔터('')를 입력할 때까지 영어단어-뜻을 입력받아 단어장을 만들고, 입력이 끝나면 단어테스트를 실시하는 프로그램을 만들어보자
  ##리스트에 입력된 순서대로 테스트를 실시하고, 맞은 개수를 계산한다.
  ##테스트가 끝나면 맞은 점수를 출력한다.

voca={}
while True:
    eng = input('eng: ')
    if eng=='':
        break
    kor = input('kor: ')
    voca[eng]=kor
print(voca)

score=0
for s in voca.keys():
    kor =input(s+':')
    if kor == voca[s]:
        score+=1

print(f'맞은 개수/전체 단어수: {score} / {len(voca.values())}')
```
