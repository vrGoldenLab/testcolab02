```python
'''
문제 : 
1) 변수는 3개 사용한다. (합계 = 0, 시작값 = 1, 끝값 = 11)
2) 시작값에서 끝값-1까지 인덱스 값을 바꾸면서 반복한다.
3) 합계값에 인덱스를 더한다.
4) 반복이 끝나면 '-'값을 50번 출력한다.
5) 합계와 인덱스 값을 출력한다.
==================================================================================
변수 이해 - test_sum : 
    >> test_sum 값은 0. 여기에 앞으로 값이 더해져나갈 것이다. 이 변수는 전체 합계를 의미한다.

난관1 : 문제부터 이해가 안갔다.
    1) 이해됐음.
    2) 어렵다. (이해완료)
        2-1) 시작값에서 끝값-1까지 라는것은 무슨 의미인가?
            >> for 변수 in 반복횟수
            >> 1부터 11까지니까 1부터 10까지 있다. 횟수로 따지면 10번이지만, 1을 10번 더하는게 아니다.
            >> range(1, 11). 1부터 10까지의 수가 순차적으로 x에 들어가게 된다. 10번의 연산을 1~10까지의 숫자로 바뀌어가는 x가 순차적으로 수행한다.
        2-2) 인덱스 값을 바꾸면서 반복한다는건?
            >> '인덱스'는 x를 의미하고, '인덱스 값'은 x에 들어가는 값을 의미한다.
            >> 인덱스 값인 x를 바꾸면서 반복문을 range에서 지정한대로 수행한다.
    3) 합계값에 인덱스를 어떻게 더하는가? 인덱스를 더하란게 무슨 의미인가? 합계값과 무슨 상관인가? (이해완료)
        >> 인덱스를 더하는 것의 의미? : 앞서 합계값 변수로 지정해둔 test_sum 에 인덱스인 x를 더하도록 지시한다. 그럼 0에 인덱스 x의 인덱스값이 들어간다.       
        >> 인덱스를 더하는 방법? : x도 그냥 변수다. test_sum에 x를 더해주고나면 그 값이 결과합계값이다. 
        >> 근데 변수를 또 선언할 수는 없으니, test_sum이 자기를 써서 연산을 수행했더니 자기가 수가 바뀌는 꼴이되게 해야한다.
        >> test_sum += x 는 이에 적합한 식.
    4) 이해됐음.
    5) 앞의 사항들이 이해가 안가서 이것도 덩달아 이해가 안됨. (이해완료)
        >> 합계는 test_sum 을 출력시키면 됨. 
        >> 인덱스 값을 출력시키라는 것은 range(1,11) 하면 너무 길어지니까, 몇부터 몇까지의 숫자가 있는지 시작과 끝만 알려주는 식으로 강사님은 수행하셨다.
'''

# 아래는 강사님의 코드

test_sum = 0
a = 1
b = 11
    # 말그대로 변수를 3개 쓰고 시작했다. 합계와 시작값, 끝값. 문제 그대로.
for x in range(a,b) :
    test_sum += x                           # += 연산자 : 왼쪽 변수에 오른쪽 값을 더하고 그 결과를 왼쪽 변수에 할당한다.
    print ("x:", x, "sum:", test_sum)
        # 이 함수 자체가 무슨 의미를 갖는 것일까?
            # 반복문을 수행하는 매번 각 횟수마다 인덱스 값인 x변수와 결과값인 test_sum를 산출하게 하는것이다. 어떻게 계산되고있는지 보려고.
        # x 변수의 for문은 a부터 b까지의 정수를 test_sum += 
print('-' * 50)
    #'-'는 실행표에서 답과 연산을 구별하기 위해 선을 긋기 위해 쓰셨다. 그냥 비주얼용인 것이다. 뭔가 특별한 트릭이 아니다.
print(a, ' 부터 ', b-1, ' 까지 더한 총합은 ', test_sum, '입니다')
    #b-1은 리스트에 그 끝값이 들어가지 않기 때문에 한 일.
```
