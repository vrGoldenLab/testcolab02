### round 함수를 사용한 성적계산 프로그램

```python
# 성적계산 프로그램 (round 함수 사용하기)

name = input("이름은 입력하세요: ")
kor = float(input("국어 성적을 입력하세요: "))
eng = float(input("영어 성적을 입력하세요: "))
math = float(input("수학 성적을 입력하세요: "))

print(divider)

print(f"{name}님의 평균 성적은 {round(((kor+eng+math)/3), 2)} 입니다.")
    #round(점수, 2) 이 부분은 점수를 소수점 둘째자리까지 표현한다는 뜻입니다!
```

## 퀴즈2 : 화씨 → 섭씨 변환기

```python
# 퀴즈 : 화씨 > 섭씨 변환기

# 더 심화로 해보고자 한다면, 화씨>섭씨, 섭씨>화씨 둘중 하나를 유저가 고르게 한 뒤 그 수식의 연산을 실시해주는 프로그램을 만들어볼 수 있다.

print("화씨에서 섭씨로 변환하는 수식은, '섭씨 = (화씨-32) * 5/9'이다.")

print(divider)

F = input('화씨온도: ')
print(f'섭씨온도: {(int(F)-int(32))*int(5)/int(9)}')
```

```python
# 화씨 섭씨 계산기 심화

divider = '-' * 50

print('''연산하고자 하는 항목의 알파벳명을 입력하세요.
C. 화씨로 섭씨 구하기.
F. 섭씨로 화씨 구하기.
''')

order = input()

if order == 'C' : 
    print("화씨에서 섭씨로 변환하는 수식은, '화씨 = (섭씨-32) * 5/9'이다.")
    print(divider)
    Faren = input('화씨온도: ')
    print(f'섭씨온도: {(float(Faren)-float(32))*float(5)/float(9)}')

elif order == 'F' : 
    print("섭씨에서 화씨로 변환하는 수식은, '섭씨 = (화씨*9/5) +32'이다.")
    print(divider)
    C = input('섭씨온도: ')
    print('화씨온도: ', (float(C)*9/5) + 32)

else : 
    print("틀린 입력입니다.")
```
