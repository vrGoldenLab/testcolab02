```python
ame_price=2000
cafe_price=3000
cafu_price=3500
ame=int(input('아메리카노 판매 갯수:'))
cafe=int(input('카페라떼 판매 갯수:'))
cafu=int(input('카푸치노 판매 갯수:'))
total_income=((ame*ame_price)+(cafe*cafe_price)+(cafu*cafu_price))
print(f'총 매출은 {total_income}원 입니다')
```
