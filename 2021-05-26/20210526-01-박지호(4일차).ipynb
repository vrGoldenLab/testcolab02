```python
#리스트 정렬 > sort
#리스트 원본을 정렬한다. 
#리스명.sort() --> 오른차순 정렬 /// 리스트명.sort(reverse=True) -->내림차순 정렬

#리스트 원본은 건드리지 않고 정렬된 새로운 리스트를 반환한다.
#sorted(리스트명)-->오름차순 정렬 /// sorted(리스트명.reverse=True)-->내림차순 정렬

#원본을 건드리지 않고 정렬된 새로운 리스트 반환하기
fruits= ['사과', '오렌지', '포도', '수박', '바나나']
sorted_fruits = sorted(fruits)
print("원본:",fruits)
print("정렬:", sorted_fruits)
```

```python
# random.choice(리스트명) > random.choice(리스트명)
import random
wise_saying = []
wise_saying.append('미래는 현재 우리가 무엇을 하고 있는가에 달려있다. - 마히트마 간디')
wise_saying.append('교육은 세상을 바꿀 수 있는 가장 강력힌 무기다 - 넬슨 만델리')
wise_saying.append('성공하기 위한 가장 확실한 방법은 한번 더 시도해 보는 것이다 - 토마스에디슨')
wise_saying.append('가장 불만에 가득찬 고객은 가장 위대한 배움의 원천이다 _ 빌 게이츠')
wise_saying.append('영원히 살 것처럼 꿈꾸고 오믈 죽을 것처럼 살아라 - 제임스딘')
wise_saying.append('경쟁우위거 없다면 경쟁하지 마라 - 잭윌치')
wise_saying.append('삶에 실수는 없고, 배움만이 있을 뿐이다. - 지아드 압델 누어')
wise_saying.append('어제와 똑같은 삶을 살면서 다른 미래를 기대하는 것은 정신병의 초기증세이다 -아인슈타인')
wise_saying.append('문제점을 찾지 말고 해결책을 찾으라 - 헨리포드')
wise_saying.append('실수를 해보지 않은 사람은 한번도 새로운 일을 시도해보지 않았던 사람이다 -아인슈타인')
print(random.choice(wise_saying))
print(random.sample(wise_saying,2))
```

```python
#random 모듈
radint(시작값, 끝값)-시작값,끝값 사이의 랜덤한 정수 반환(끝값 포함)
randrange(시작값, 끝값)-시작값, 끝값 사이의 랜덤한 정수 반환(끝값 미포함)
randrange(끝값)-인자가 1개인 경우 0~끝값 사이의 랜덤한 정수 반환(끝값 미포함)
random()-0.0~0.1 사이의 실수를 반환 (1.0미포함)
uniform(시작값, 끝값)-시작값~끝값 사이의 실수를 반환(끝값 미포함)
choice(sequence)-시퀸스 데어터 내에서 무작위로 하나의 원소 추출
sample(sequence,n)-시퀸스 데이터 내에서 무작위로 unique한 n개의 원소를 추출
shuffle(sequence)-시퀸스 데이터를 무작위로 랜덤하게 섞는다.
```
